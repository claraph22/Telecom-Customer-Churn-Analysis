## Conceptual Problems




1. Latar Belakang Adanya Bagging dan Cara Kerja Bagging

**Latar Belakang:**


Bagging (Bootstrap Aggregating) adalah teknik yang dikembangkan untuk meningkatkan akurasi model prediksi dengan mengurangi varians. Latar belakang adanya bagging adalah untuk mengatasi overfitting pada model machine learning, terutama pada model yang kompleks seperti pohon keputusan.

**Cara Kerja Bagging:**


1. Membuat beberapa subset data pelatihan dengan pengambilan sampel acak (dengan pengembalian) dari dataset asli.
2. Melatih model yang sama pada setiap subset.
3. Menggabungkan hasil prediksi dari semua model dengan cara rata-rata (untuk regresi) atau voting mayoritas (untuk klasifikasi).

 2. Perbedaan Cara Kerja Algoritma Random Forest dengan Algoritma Boosting

**Random Forest:**


- Menggunakan metode bagging.
- Membuat banyak pohon keputusan dari subset data yang diambil secara acak.
- Menggabungkan prediksi dari semua pohon dengan rata-rata atau voting mayoritas.
- Setiap pohon dibuat secara independen dari pohon lain.

**Boosting (contoh: AdaBoost):**


- Membuat model secara berurutan.
- Setiap model baru difokuskan untuk memperbaiki kesalahan dari model sebelumnya.
- Data yang salah diprediksi oleh model sebelumnya diberi bobot lebih besar.
- Menggabungkan semua model dengan cara menimbang prediksi berdasarkan akurasi masing-masing model.

 3. Cross Validation



Cross Validation adalah teknik untuk mengevaluasi kinerja model machine learning dengan cara membagi data pelatihan menjadi beberapa subset. Model dilatih dan diuji pada subset yang berbeda secara bergantian. Metode yang paling umum adalah k-fold cross validation, di mana data dibagi menjadi k subset (fold), dan proses pelatihan serta pengujian dilakukan k kali, masing-masing dengan satu subset sebagai data uji dan sisanya sebagai data latih. Hasil akhirnya adalah rata-rata kinerja dari semua iterasi. Teknik ini membantu mengurangi bias dan varians dalam evaluasi model.

Encoding fitur diperlukan untuk mengubah data kategorikal menjadi data numerik.
Hal ini penting karena sebagian besar algoritma pembelajaran mesin hanya dapat bekerja dengan data numerik.
Selain itu, encoding fitur dapat membantu untuk mengatasi masalah overfitting dan underfitting.

```python
# prompt: buat pipeline OneHotEncoder

from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Define the categorical features
cat_features = ['SeniorCitizen', 'Partner', 'Dependents', 'InternetService', 'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod']

# Define the pipeline
ohe_pipeline = Pipeline([
    ('ohe', OneHotEncoder(handle_unknown='ignore', sparse=False))
])

# Create the preprocessor
preprocessor = ColumnTransformer([
    ('ohe', ohe_pipeline, cat_features)
])

# Fit the preprocessor on the training data
preprocessor.fit(X_train)

# Transform the training and testing data
X_train_encoded = preprocessor.transform(X_train)
X_test_encoded = preprocessor.transform(X_test)
```

```python
# Definisikan pipeline untuk data kategori
categorical_pipeline = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse=False))
])

# Tentukan kolom kategori
categorical_features = X_train_cat.columns

# Membuat ColumnTransformer yang akan menerapkan pipeline untuk kolom kategori
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_pipeline, categorical_features)
    ],
    remainder='passthrough'  # This keeps the non-specified columns unchanged
)
```

## Kelebihan & Kekurangan pada Algoritma

| Algoritma                      | Kelebihan                                                                                  | Kekurangan                                                                                       |
|--------------------------------|--------------------------------------------------------------------------------------------|-------------------------------------------------------------------------------------------------|
| **K-Nearest Neighbors (KNN)**  | - Simplicity: Mudah dipahami dan diimplementasikan                                         | - Computationally Expensive: Memerlukan banyak waktu dan memori untuk prediksi                   |
|                                | - Non-parametric: Tidak membuat asumsi tentang distribusi data                             | - Sensitive to Irrelevant Features: Kinerja dapat menurun jika terdapat fitur yang tidak relevan |
|                                | - Adaptability: Dapat menangani kasus non-linear dengan baik                               | - Curse of Dimensionality: Kinerja menurun jika jumlah fitur sangat besar                        |
| **Support Vector Classifier (SVC)** | - Effective in High Dimensions: Bekerja dengan baik ketika jumlah fitur lebih besar dari jumlah sampel | - Computationally Intensive: Latihan bisa sangat lambat pada dataset besar                       |
|                                | - Robustness: Toleran terhadap outlier dengan menggunakan kernel trick                     | - Parameter Tuning: Memerlukan pemilihan dan penyetelan parameter yang tepat                     |
|                                | - Versatile: Mendukung berbagai kernel functions (linear, polynomial, RBF)                 | - Interpretability: Sulit untuk menginterpretasikan model hasil pelatihan                        |
| **Decision Tree Classifier**   | - Easy to Understand and Interpret: Struktur pohon keputusan intuitif dan mudah diinterpretasikan | - Prone to Overfitting: Cenderung overfitting terutama pada dataset yang kecil                   |
|                                | - No Need for Data Normalization: Tidak memerlukan normalisasi atau scaling data           | - Unstable: Perubahan kecil pada data dapat menghasilkan pohon yang sangat berbeda               |
|                                | - Handling of Non-linear Relationships: Mampu menangani hubungan non-linear dalam data     | - Bias: Memiliki kecenderungan bias jika pohon tidak dalam (terlalu dangkal)                     |
| **Random Forest Classifier**   | - Reduced Overfitting: Mengurangi overfitting dengan menggabungkan prediksi dari banyak pohon keputusan | - Complexity: Sulit diinterpretasikan dibandingkan dengan pohon keputusan tunggal                |
|                                | - Robustness: Stabil dan tahan terhadap noise dan outlier                                  | - Computationally Intensive: Memerlukan lebih banyak sumber daya komputasi                       |
|                                | - Feature Importance: Memberikan estimasi pentingnya setiap fitur dalam prediksi           | - Not Always Optimal: Hasilnya mungkin tidak sebaik metode boosting pada beberapa kasus          |
| **Gradient Boosting Classifier** | - High Performance: Sering kali memberikan kinerja terbaik dalam banyak tugas klasifikasi | - Slow Training: Proses pelatihan bisa sangat lambat                                             |
|                                | - Flexibility: Dapat digunakan dengan berbagai fungsi kerugian                             | - Parameter Tuning: Memerlukan penyetelan parameter yang cermat                                  |
|                                | - Handling of Missing Data: Mampu menangani data yang hilang dengan baik                   | - Overfitting: Masih bisa overfitting jika tidak dikendalikan dengan baik                        |


**Kesimpulan**


Pemilihan algoritma yang tepat tergantung pada karakteristik dataset dan kebutuhan spesifik dari masalah yang ingin diselesaikan.


- Gunakan KNN jika data sederhana dan interpretasi tidak terlalu penting.

- Pilih SVC untuk data dengan dimensi tinggi dan jika Anda membutuhkan hasil yang sangat akurat.


- Pilih Decision Tree untuk interpretasi yang mudah dan jika ingin hasil yang cepat.

- Pilih Random Forest untuk mengurangi overfitting dan mendapatkan model yang stabil.

- Gunakan Gradient Boosting jika kinerja model yang sangat tinggi adalah prioritas dan Anda bersedia menghabiskan waktu untuk tuning parameter.

```python
# Import Libraries
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV
import numpy as np

# Define Parameters for RandomizedSearchCV
n_neighbors = [int(x) for x in np.linspace(3, 15, 7)]  # Use parentheses for calling the function np.linspace
weights = ['uniform', 'distance']
algorithm = ['auto', 'ball_tree', 'kd_tree']
leaf_size = [int(x) for x in np.linspace(20, 50, 3)]
random_search_params_knn = {
    'knn__n_neighbors': n_neighbors,  # Note the 'knn__' prefix
    'knn__weights': weights,
    'knn__algorithm': algorithm,
    'knn__leaf_size': leaf_size
}

# Define Pipeline
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Preprocessing step: scaling
    ('knn', KNeighborsClassifier())  # Model: KNN
])

# Perform Randomized Search
knn_randomcv = RandomizedSearchCV(estimator=pipeline,
                                  param_distributions=random_search_params_knn,
                                  cv=5,
                                  random_state=46,
                                  n_jobs=-1,
                                  scoring='f1')
knn_randomcv.fit(X_train_final, y_train)
```

## Bussiness Strategy

 **1. Targetkan Pelanggan Baru:**
- Fokus pada menarik pelanggan baru, terutama yang berusia lebih muda, untuk meningkatkan basis pelanggan dan mengurangi tingkat churn.

 **2. Promosikan Layanan Tambahan:**
- Tingkatkan upaya pemasaran untuk mempromosikan layanan tambahan seperti keamanan online, backup online, dan perlindungan perangkat untuk meningkatkan pendapatan per pelanggan.

**3. Pertahankan Pelanggan Lama:**
- Berikan insentif kepada pelanggan yang sudah ada untuk tetap berlangganan, seperti diskon atau paket khusus.
- Identifikasi pelanggan yang berisiko churn dan berikan penawaran khusus untuk mempertahankan mereka.

**4. Optimalkan Kontrak Pelanggan:**
- Dorong pelanggan untuk beralih ke kontrak yang lebih lama, karena pelanggan dengan kontrak yang lebih lama cenderung bertahan lebih lama.
- Pertimbangkan untuk menawarkan diskon atau insentif lain untuk mendorong pelanggan agar beralih ke kontrak yang lebih lama.

**5. Pantau dan Evaluasi Kinerja:**
- Terus pantau tingkat churn dan metrik kinerja lainnya untuk menilai efektivitas strategi bisnis.
- Gunakan data untuk mengidentifikasi area di mana perbaikan dapat dilakukan dan sesuaikan strategi bisnis sesuai kebutuhan.

Tindakan yang Direkomendasikan
Mengatasi Overfitting:
- **Tuning Hyperparameters**: Sesuaikan parameter model untuk menemukan konfigurasi yang lebih umum.
- **Regularisasi**: Terapkan teknik regularisasi seperti L1 atau L2 untuk mengurangi kompleksitas model.
- **Cross-Validation**: Gunakan teknik cross-validation untuk mengevaluasi model dengan lebih baik dan menghindari overfitting.

Peningkatan Precision:
- Fokus pada peningkatan precision, terutama untuk model KNN dan Random Forest, dengan mengurangi jumlah false positives.
- Lakukan analisis kesalahan untuk memahami pola kesalahan dan memperbaiki model.

Balancing Data:
- Pertimbangkan penggunaan teknik balancing data seperti SMOTE (Synthetic Minority Over-sampling Technique) untuk menangani ketidakseimbangan kelas.

Evaluasi Lanjutan:
- Gunakan metrik tambahan seperti precision, recall, F1-score, dan ROC-AUC untuk mendapatkan gambaran yang lebih lengkap tentang performa model.
- Lakukan analisis mendalam terhadap kesalahan prediksi untuk meningkatkan akurasi prediksi model.
